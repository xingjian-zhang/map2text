```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px

data_files = {
    "persona": {
        "text": "persona.tsv",
        "embedding": "persona.npz",
    },
    "cs-research-idea": {
        "text": "massw.tsv",
        "embedding": "key_ideas.npz",
    },
    "cs-research-context": {
        "text": "massw.tsv",
        "embedding": "context.npz",
    },
    "prompt-log": {
        "text": "prompt_log.tsv",
        "embedding": "prompt_log.npz",
    },
    "red_team_attempts": {
        "text": "red_team_attempts.tsv",
        "embedding": "red_team_attempts.npz",
    },
}

def load_data(dataset_name):
    data_file = data_files[dataset_name]
    text_path = f"../data/{data_file['text']}"
    embedding_path = f"../data/{data_file['embedding']}"

    df = pd.read_csv(text_path, sep="\t")
    embeddings = np.load(embedding_path)
    return df, embeddings["low_dim_embeddings"]
```

```python
# Get basic stats
# 1. Number of rows in low_dim_embeddings
# 2. Average length of text (#tokens)

import tiktoken

encoding = tiktoken.encoding_for_model("gpt-4o")
def count_tokens(text):
    num_tokens = len(encoding.encode(str(text)))
    return num_tokens

df_red_team_attempts, embeddings_red_team_attempts = load_data("red_team_attempts")
print(f"Number of rows in low_dim_embeddings: {embeddings_red_team_attempts.shape[0]}")
print(f"Average length of text: {df_red_team_attempts['content'].apply(count_tokens).mean():.2f} tokens")

df_persona, embeddings_persona = load_data("persona")
print(f"Number of rows in low_dim_embeddings: {embeddings_persona.shape[0]}")
print(f"Average length of text: {df_persona['persona'].apply(count_tokens).mean():.2f} tokens")

df_idea, embeddings_idea = load_data("cs-research-idea")
print(f"Number of rows in low_dim_embeddings: {embeddings_idea.shape[0]}")
print(f"Average length of text: {df_idea['key_idea'].apply(count_tokens).mean():.2f} tokens")

df_context, embeddings_context = load_data("cs-research-context")
print(f"Number of rows in low_dim_embeddings: {embeddings_context.shape[0]}")
print(f"Average length of text: {df_context['context'].apply(count_tokens).mean():.2f} tokens")

```

```python

df_persona, embeddings_persona = load_data("persona")
df_persona["x"] = embeddings_persona[:, 0]
df_persona["y"] = embeddings_persona[:, 1]

fig = px.scatter(
    data_frame=df_persona,
    x="x",
    y="y",
    hover_data=["persona"],
    opacity=0.5,
)
fig.update_traces(marker=dict(size=2, opacity=0.2, color="#1f77b4"))
fig.update_layout(showlegend=False)
fig.update_xaxes(visible=False, range=[min(embeddings_persona[:, 0]) + 1, max(embeddings_persona[:, 0])])
fig.update_yaxes(visible=False, range=[min(embeddings_persona[:, 1]), max(embeddings_persona[:, 1])])
fig.update_layout(width=1000, height=1000)
fig.update_layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)'
)
fig.show()
fig.write_image("persona.png", width=1000, height=1000, scale=2)
```
