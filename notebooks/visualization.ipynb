```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data_files = {
    "persona": {
        "text": "persona.tsv",
        "embedding": "persona.npz",
    },
    "cs-research-idea": {
        "text": "massw.tsv",
        "embedding": "key_ideas.npz",
    },
    "cs-research-context": {
        "text": "massw.tsv",
        "embedding": "context.npz",
    },
    "prompt-log": {
        "text": "prompt_log.tsv",
        "embedding": "prompt_log.npz",
    },
    "red_team_attempts": {
        "text": "red_team_attempts.tsv",
        "embedding": "red_team_attempts.npz",
    },
}

def load_data(dataset_name):
    data_file = data_files[dataset_name]
    text_path = f"../data/{data_file['text']}"
    embedding_path = f"../data/{data_file['embedding']}"

    df = pd.read_csv(text_path, sep="\t")
    embeddings = np.load(embedding_path)
    return df, embeddings["low_dim_embeddings"]
```

```python
import plotly.express as px

df_red_team_attempts, embeddings_red_team_attempts = load_data("red_team_attempts")

df_red_team_attempts["x"] = embeddings_red_team_attempts[:, 0]
df_red_team_attempts["y"] = embeddings_red_team_attempts[:, 1]

# Hover text
fig = px.scatter(
    data_frame=df_red_team_attempts,
    x="x",
    y="y",
    hover_data=["content"],
    opacity=0.5,
)

# Add hover text
fig.update_traces(marker=dict(size=2, opacity=0.2, color="#1f77b4"))
fig.update_layout(showlegend=False)
fig.update_xaxes(visible=False, range=[min(embeddings_red_team_attempts[:, 0]) + 1, max(embeddings_red_team_attempts[:, 0])])
fig.update_yaxes(visible=False, range=[min(embeddings_red_team_attempts[:, 1]), max(embeddings_red_team_attempts[:, 1])])
fig.update_layout(width=1000, height=1000)
fig.update_layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)'
)
fig.show()

```
